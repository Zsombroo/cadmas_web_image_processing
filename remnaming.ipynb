```python
import pandas as pd
```

```python
df = pd.read_csv("nevek.csv")
df = df[['jpg_nev', 'termekkod', 'sorrend']]
df = df.dropna(axis=0)
df = df.loc[df['jpg_nev'].str.startswith('DSC')]
df
```

```python
arr = df.to_numpy()
arr[:5]
```

```python
nevek = []
for i in arr:
    nevek.append([i[0], f'{i[1]}_altpic_{i[2]}.jpg'])
```

```python
import os
path = 'negyzetesitett'

bads = []

for i in nevek:
    try:
        os.rename(os.path.join(path, i[0]), os.path.join(path, i[1]))
        print(i)
    except Exception:
        bads.append(i)
print(len(bads))
```

```python
for name in os.listdir(path):
    if '-' in name:
        new_name = '_'.join(name.split('-'))
        try:
            os.rename(os.path.join(path, name), os.path.join(path, new_name))
        except FileExistsError as e:
            print(e)
```
